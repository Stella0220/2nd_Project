```python
# 육아휴직제도사용실적현황.csv 파일 읽기
import pandas as pd
import numpy as np

df=pd.read_csv('육아휴직제도사용실적현황.csv', low_memory=False, encoding='cp949')
df
```

```python
columns2=['사업체특성별(1)','사업체특성별(2)']

for i in range(6):
    a=2017.1
    b=2017.3
    a+=i
    columns2.append(str(a))
    b+=i
    columns2.append(str(b))           
    
print(columns2) # 확인용 

df2=df[columns2]
df2.head()
```

```python
# [csv파일 CLEANSING 및 FIXING]
# 위의 자료에서 '사용실적여부-비율(%)' 정보 열만 추출
# 연도별 '사용실적여부' 열의 규칙: 각 연도별 .1과 .3의 열 (예) 2017.1, 2017.3, 2018.1, 2018.3...2022.3

columns3=['사업체특성별(1)','사업체특성별(2)']

for i in range(6):
    a=2017
    b=2017
    a+=i
    columns3.append(str(a)+' 있다')
    b+=i
    columns3.append(str(b)+' 없다')           
    
print(columns3) # 확인용 

# 컬럼명 변경
df2.columns = columns3
df2

# 인덱스: '사업체특성별(1)','사업체특성별(2)'를 인덱스로 변경
df2.set_index(['사업체특성별(1)','사업체특성별(2)'], inplace=True)
df2.head()  
```

```python
# 원본 데이터프레임에서 첫 3행 삭제
df3 = df2.drop(df2.index[:4],axis=0)
df3=df3.astype(float)
df3.head()
```

```python
# # 인덱스가 '종사자규모별'인 행 선택
size = df3.xs('종사자규모별',level=0)

size=size.reset_index()
col=[['','2017','2017','2018','2018','2019','2019','2020','2020','2021','2021','2022','2022'],\
     ['사업체특성별(2)','사용 경험 O','사용 경험 X','사용 경험 O','사용 경험 X','사용 경험 O','사용 경험 X',\
      '사용 경험 O','사용 경험 X','사용 경험 O','사용 경험 X','사용 경험 O','사용 경험 X']]
size.columns = col

size=size.rename(columns={'사업체특성별(2)':'기업규모별'}).T

size=size.rename(columns=size.iloc[0]).reset_index()
size=size.drop([0],axis=0)
size=size.rename(columns={'level_0':'연도','level_1':'육아휴직 사용 유무'}).reset_index().drop('index',axis=1)
size=size.set_index(keys='연도')

# 5~9인 규모 기업 육아휴직 현황
size1=size.iloc[:,:2]
size1

size1=size1.reset_index().drop('연도',axis=1).T
size1=size1.rename(columns=size1.iloc[0]).drop(index='육아휴직 사용 유무')

size1
a=size1.iloc[:,0:2]
b=size1.iloc[:,2:4]
c=size1.iloc[:,4:6]
d=size1.iloc[:,6:8]
e=size1.iloc[:,8:10]
f=size1.iloc[:,10:12]

corp59=pd.concat([a,b,c,d,e,f],axis=0)

index=['2017','2018','2019','2020','2021','2022']
corp59.index=index
corp59.index.name='연도'

corp59

size
```
